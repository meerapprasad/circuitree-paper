```python
import numpy as np
import pandas as pd
import networkx as nx
from pathlib import Path
```

```python
%matplotlib widget
```

```python
gml = Path("../data/aws/mcts/231005-03-12-05_5tf_lockfree/backups/tree-c3823ac2-be81-4410-9fea-e29b33bc2f44_2023-10-07_12-11-39.gml.gz")
```

```python
graph = nx.read_gml(gml)
```

```python
terminal_nodes = list(n for n in graph.nodes if n[0] == "*")
```

```python
top_20_visits = sorted(terminal_nodes, key=lambda n: graph.nodes[n]["reward"], reverse=True)
```

```python
top_20_visits
```

```python
for n in top_20_visits:
    print(graph.nodes[n]["visits"], graph.nodes[n]["reward"])
```

```python
topq = sorted(terminal_nodes, key=lambda n: graph.nodes[n]["reward"] / max(1, graph.nodes[n]["visits"]), reverse=True)
```

```python
for i in range(100):
    n = topq[i]
    r = graph.nodes[n]["reward"]
    v = graph.nodes[n]["visits"]
    print(n)
    print(f"\t{r:.0f} / {v}")
```
